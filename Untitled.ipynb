```python
import pandas as pd

tickets_df = pd.read_csv("tickets.csv")
tickets_df.head()
```

```python
import re

def clean_text(text):
    text = text.lower()
    text = re.sub(r"[^a-zA-Z0-9\s]", "", text)
    return text.strip()

tickets_df['clean_text'] = tickets_df['ticket_text'].apply(clean_text)
```

```python
from transformers import AutoTokenizer

# Pretrained transformer model tokenizer
tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased")

# Tokenize text
tokens = tokenizer(
    tickets_df['clean_text'].tolist(),
    padding=True,
    truncation=True,
    max_length=128,
    return_tensors="tf"
)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
tickets_df['label'] = le.fit_transform(tickets_df['category'])
labels = tickets_df['label'].tolist()
```

```python
import tensorflow as tf

# Assume tokens is a dictionary of lists/arrays
# Convert all inputs to tensors first
tokens_tf = {k: tf.constant(v) for k, v in tokens.items()}
labels_tf = tf.constant(labels)

# Build dataset from tensors
dataset = tf.data.Dataset.from_tensor_slices((tokens_tf, labels_tf))
dataset = dataset.shuffle(buffer_size=1024).batch(8)

# Check one batch
for batch in dataset.take(1):
    x, y = batch
    print(x)
    print(y)
```

```python
from transformers import TFAutoModelForSequenceClassification, AutoTokenizer

MODEL_NAME = "distilbert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)

model = TFAutoModelForSequenceClassification.from_pretrained(
    MODEL_NAME,
    from_pt=True,  # <-- tells library to convert PyTorch weights to TF
    num_labels=len(le.classes_)
)

```

```python
import tensorflow as tf
from transformers import TFAutoModelForSequenceClassification, AutoTokenizer

# Model & tokenizer
MODEL_NAME = "distilbert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
model = TFAutoModelForSequenceClassification.from_pretrained(
    MODEL_NAME,
    from_pt=True,
    num_labels=len(le.classes_)
)

# Optimizer and loss
optimizer = tf.keras.optimizers.Adam(learning_rate=5e-5)
loss_fn = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True)

# Example training loop
for epoch in range(3):
    print(f"Epoch {epoch+1}")
    for batch in dataset:  # Assume dataset is a tf.data.Dataset
        inputs = {k: v for k, v in batch[0].items()}  # tokens
        labels = batch[1]
        with tf.GradientTape() as tape:
            outputs = model(**inputs, training=True)
            logits = outputs.logits
            loss = loss_fn(labels, logits)
        gradients = tape.gradient(loss, model.trainable_variables)
        optimizer.apply_gradients(zip(gradients, model.trainable_variables))
    print(f"Loss: {loss.numpy()}")
```
